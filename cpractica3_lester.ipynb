```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = 'train.csv'
df = pd.read_csv(data)
```

```python
print(df.head())
```

```python
print(df.info()) # Informaci√≥n sobre las columnas y tipos de datos
```

```python
print(df.describe())
```

```python
product_freq = df['Product ID'].value_counts()
print(product_freq)
```

```python
top_5_products = df.groupby('Product Name')['Sales'].sum().nlargest(5)
top_5_products.plot(kind='bar',title='Top 5 productos mas vendidos')
plt.ylabel('Ventas Producto')
plt.show()
```

```python
plt.hist(df['Sales'],bins=20,color='blue')
plt.title('Histograma de ventas')
plt.xlabel('Ventas')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Volver a generar el DataFrame 'top_5_ciudades' y crear el gr√°fico de pastel de nuevo

# Seleccionar las 5 ciudades con m√°s ventas
top_5_ciudades = df.nlargest(5, 'Sales')

# Crear un gr√°fico de pastel para visualizar las ventas de las 5 ciudades con m√°s ventas
plt.figure(figsize=(8, 8))
plt.pie(top_5_ciudades['Sales'], labels=top_5_ciudades['City'], autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'coral', 'orange', 'pink'])
plt.title('Distribucion de ventas en las 5 Ciudades con M√°s Ventas')
plt.axis('equal')  # Asegura que el gr√°fico de pastel sea circular
plt.tight_layout()

# Mostrar el gr√°fico
plt.show()
```
